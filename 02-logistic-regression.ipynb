## Pass/Fail Classification
if >= 50 pass

```python
rng = np.random.default_rng(seed=42)
X = rng.integers(0, 101, size=500)
# Print First 100 row
X[:100]
```

```python
y = np.where(X >= 50, 1, 0)
y[:100]
```

```python
plt.scatter(X, y, alpha=0.1, cmap='viridis' )
plt.axvline(x=50, color='red', linestyle='--', linewidth=1.5, label='Boundary (X=50)')
plt.show()
```

```python
model = LogisticRegression()
model.fit(X, y, epoch=10000, _lamda=0.006)
```

```python
x_test = np.array([1, 20, 30, 66, 49, 50, 51, 48, 52, 53, 54, 47])
y_pred = model.predict(x_test)
y_pred = np.where(y_pred >= 0.5, 1, 0)
y_pred
```

```python
y_true = np.where(x_test >= 50, 1, 0)
plt.scatter(x_test, y_true, color='blue', label='True', alpha=0.7, edgecolors='none', s=50)
plt.scatter(x_test, y_pred + 0.02, color='red', label='Pred', alpha=0.5, edgecolors='none', s=20)

plt.xlabel("Student's scores")
plt.ylabel('Pass(1) / Fail(0)')
plt.ylim(-0.1, 1.1)
plt.show()
```

```python
m_test = len(x_test)
errors = 0
for i in range(m_test):
    if y_pred[i] != y_true[i]:
        print("X when error:", x_test[i], "| Pred:", y_pred[i])
        errors += 1
corrects = m_test - errors
print("Percentage of correctness: %", 100 * (corrects / m_test))
```

```python
y_pred = model.predict(X)
y_pred = np.where(y_pred>=0.5, 1, 0)
m = len(X)
errors = 0
for i in range(m):
    if y_pred[i] != y[i]:
        print("X when error:", X[i], "| Pred:", y_pred[i])
        errors += 1
corrects = m - errors
print("Percentage of correctness: %", 100 * (corrects / m))
```
