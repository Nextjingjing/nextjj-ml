## Linear Regression
Linear regression เป็น supervised learning algorithm ซึ่งใช้ทำ Model แสดงความสัมพันธ์แบบเชิงเส้นระหว่าง 1 ตัวแปรตาม (Dependent variable / $\mathbf{y}$) และมากกว่า 1 ตัวแปรอิสระ ( Independent variable / $\vec{x}$ ) 

> ใช้ทำนายตัวเลข $\hat{y}$ จากค่า $\vec{x}$ ซึ่ง $\hat{y}$ เป็นตัวเลขจำนวนจริงอะไรก็ได้

### ตัวอย่างการใช้งานเช่น
- การทำนายราคาบบ้าน ($\mathbf{y}$) จากขนาดบ้าน, จำนวนชั้น, จำนวนห้องนอน, ... ($\vec{x}$)

## Notations
- $m$ = จำนวนข้อมูลทั้งหมด (Examples / Rows)
- $n$ = จำนวนตัวแปรอิสระหรือฟีเจอร์ (Features / Columns)
- $\mathbf{w}$ = Weights (ค่าสัมประสิทธิ์/ความชัน) มีขนาดตามจำนวนฟีเจอร์ ($n$)
- $b$ = Bias (ค่าคงที่/จุดตัดแกน Y)
- $\vec{x}$ = เวกเตอร์ของฟีเจอร์สำหรับข้อมูล 1 ตัวอย่าง
- $\vec{x}^{(i)}$ = ข้อมูลของ x แถวที่ i
- $\vec{y}^{(i)}$ = ข้อมูลของ y แถวที่ i
- $x_i$ = Features ที่ i ของ x

## สูตรที่ใช้งาน
$$\hat{y} = w_1 x_1 + w_2 x_2 + \dots + w_n x_n + b$$
ซึ่งเขียนเป็น Vector ได้ว่า
$$\hat{y} = \vec{w} \cdot \vec{x} + b$$

### Cost Function
$$J(\mathbf{w}, b) = \frac{1}{2m} \sum_{i=1}^{m} \left( \hat{y}^{(i)} - y^{(i)} \right)^2$$
ค่านี้ใช้บอกว่า Model เราผิดพลาดมากแค่ไหน ซึ่งเราควรทำให้ Cost Function น้อยที่สุดเท่าที่เป็นไปได้

### Gradient decent
#### การหาค่า W ที่เหมาะสม
$$w_j := w_j - \alpha \frac{\partial J(\mathbf{w}, b)}{\partial w_j}$$
ซึ่งมี j = 1, 2, 3, ..., n โดยที่
$$\frac{\partial J(\mathbf{w}, b)}{\partial w_j} := \frac{1}{m} \sum_{i=1}^{m} \left( \hat{y}^{(i)} - y^{(i)} \right) x_j^{(i)}$$

#### การหาค่า b ที่เหมาะสม
$$b := b - \alpha \frac{\partial J(\mathbf{w}, b)}{\partial b}$$
$$\frac{\partial J(\mathbf{w}, b)}{\partial b} := \frac{1}{m} \sum_{i=1}^{m} \left( \hat{y}^{(i)} - y^{(i)} \right)$$

### อัลกอริทึมหาค่า Cost Function ให้น้อยที่สุด
Repeat {
$$\begin{aligned}
w_j &:= w_j - \alpha \frac{\partial J(\mathbf{w}, b)}{\partial w_j} \\
b &:= b - \alpha \frac{\partial J(\mathbf{w}, b)}{\partial b}
\end{aligned}$$
} 

จนกว่าจะ Converge (หรือจนกว่าการเปลี่ยนแปลงของค่า Cost Function จะลดลงต่ำกว่า Threshold ที่เรากำหนดไว้) ซึ่งที่ผมทำคือกำหนด Epoch แต่จะมีการ Print Cost function ในแต่ละ Epoch ให้ดู


```python
X = np.arange(10)
y = 5 * X + 2
print("X:", X)
print("y:", y)
```

```python
model = LinearRegression()
model.fit(X, y, epoch=3000, _lamda=0.01)
print("--- Result ---")
print("w:", model.w)
print("b:", model.b)
```

```python
# Predict
x_test = np.array([1, 2, 3, 4, 5])
y_pred = model.predict(x_test)
y_pred
```

```python
fig, ax = plt.subplots()
ax.scatter(X, y, label = "Data")
ax.plot(X, model.predict(X), label = "Pred")
plt.xlabel('x label')
plt.ylabel('y label')
plt.legend()
```

```python
X = np.arange(30).reshape(-1, 3)
X
```

```python
m, n = X.shape
y = np.zeros(m)
w = np.array([5, 10 ,15])
b = 20
for i in range(m):
    y[i] = np.dot(w, X[i]) + b
y
```

```python
model = LinearRegression()
model.fit(X, y, epoch=10000, _lamda=0.001)
print("--- Result ---")
print("w:", model.w)
print("b:", model.b)
```

```python
# Data vs Prediction

y_pred = model.predict(X)
for i in range(m):
    print(f"Data: {y[i]} | Pred: {y_pred[i]}")
```
