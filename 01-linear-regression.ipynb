```python
import numpy as np
from src.models.linear_regression import LinearRegression
import matplotlib.pyplot as plt
```

```python
X = np.arange(10)
y = 5 * X + 2
print("X:", X)
print("y:", y)
```

```python
model = LinearRegression()
model.fit(X, y, epoch=3000, _lamda=0.01)
print("--- Result ---")
print("w:", model.w)
print("b:", model.b)
```

```python
# Predict
x_test = np.array([1, 2, 3, 4, 5])
y_pred = model.predict(x_test)
y_pred
```

```python
fig, ax = plt.subplots()
ax.scatter(X, y, label = "Data")
ax.plot(X, model.predict(X), label = "Pred")
plt.xlabel('x label')
plt.ylabel('y label')
plt.legend()
```

```python
X = np.arange(30).reshape(-1, 3)
X
```

```python
m, n = X.shape
y = np.zeros(m)
w = np.array([5, 10 ,15])
b = 20
for i in range(m):
    y[i] = np.dot(w, X[i]) + b
y
```

```python
model = LinearRegression()
model.fit(X, y, epoch=10000, _lamda=0.001)
print("--- Result ---")
print("w:", model.w)
print("b:", model.b)
```

```python
# Data vs Prediction

y_pred = model.predict(X)
for i in range(m):
    print(f"Data: {y[i]} | Pred: {y_pred[i]}")
```
